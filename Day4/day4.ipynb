```python
import pandas as pd
import numpy as np
```

```python
with open('day4.input', 'r') as file:
    letter_list = []
    for line in file:
        letter_list.append(list(line.strip()))
    df_xmas = (pd.DataFrame(letter_list))

df_xmas.head()
```

```python
X_locations = list(zip(*df_xmas.isin(['X']).to_numpy().nonzero()))
M_locations = list(zip(*df_xmas.isin(['M']).to_numpy().nonzero()))
A_locations = list(zip(*df_xmas.isin(['A']).to_numpy().nonzero()))
S_locations = list(zip(*df_xmas.isin(['S']).to_numpy().nonzero()))
xmas_count = 0
for X_loc in X_locations:
    x,y = X_loc
    hor_fw = [(x,y), (x,y+1), (x,y+2), (x,y+3)]
    hor_rv = [(x,y), (x,y-1), (x,y-2), (x,y-3)]
    ver_fw = [(x,y), (x+1,y), (x+2,y), (x+3,y)]
    ver_rv = [(x,y), (x-1,y), (x-2,y), (x-3,y)]
    diaup_fw = [(x,y), (x+1,y+1), (x+2,y+2), (x+3,y+3)]
    diaup_rv = [(x,y), (x-1,y-1), (x-2,y-2), (x-3,y-3)]
    diad_fw = [(x,y), (x+1,y-1), (x+2,y-2), (x+3,y-3)]
    diad_rv = [(x,y), (x-1,y+1), (x-2,y+2), (x-3,y+3)]
    search_vectors = [hor_fw,hor_rv,ver_fw,ver_rv,diaup_fw,diaup_rv,diad_fw,diad_rv]
    for vec in search_vectors:
        if vec[1] in M_locations:
            if vec[2] in A_locations:
                if vec[3] in S_locations:
                    xmas_count += 1
print (xmas_count)
```

```python
A_locations = list(zip(*df_xmas.iloc[1:-1,1:-1].isin(['A']).to_numpy().nonzero()))

mas_count = 0
for A_loc in A_locations:
    x,y = A_loc
    x,y = x+1,y+1
    MandS1 = [df_xmas.iloc[x-1,y+1], df_xmas.iloc[x+1,y-1]]
    MandS2 = [df_xmas.iloc[x-1,y-1], df_xmas.iloc[x+1,y+1]]
    if ('M' in MandS1) and ('S' in MandS1):
        if ('M' in MandS2) and ('S' in MandS2):
            mas_count += 1

print (mas_count)
```
